```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=200,)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show() #so this is without noise so whens ome data set come with noise then 
```

```python
X,y=make_moons(n_samples=200,noise=0.20)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
X,y=make_moons(n_samples=200,noise=0.10)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()#so its not  linearly seprable dataset
#so its non convex not linearly seprable
#lets test Kmeans on it!
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2)
km.fit(X)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
Centers=km.cluster_centers_
```

```python
labels=km.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.show()
#So kmeans fail on datasset which are complex or linearly seprable
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(Centers[:,0],Centers[:,1],c="magenta",s=300,marker="*")
```
